## Supermarkets/Convenience Stores per Zipcode

This notebook analyzes the convenience stores and supermarket dataset and counts how many each zipcode has.

First, we have to import pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Next, we have to create a dataframe for our dataset. I had to change the zipcodes to integer form because this dataset had them as floats instead of integers like the other datasets.

```python
stores = pd.read_csv("stores.tsv",index_col="_id",sep="\t")
stores=stores.dropna()
stores['Zip'] = stores['Zip'].astype(int)

stores.head(10)
```

Since this dataset had more zipcodes than the others, we have to filter out the zipcodes that the other datasets didn't have. Then we count how many times each zipcode appears in the dataframe, which gives us how many convenience stores and supermarkets each zipcode has.

```python
zipcodesToCount = [15212, 15210, 15206, 15226, 15216, 15217, 15219, 15208, 15201, 15233, 15213, 15207, 15205, 15204, 15224, 15214, 15222, 15220, 15203, 15221, 15232, 15211]

filteredStores = stores[stores["Zip"].isin(zipcodesToCount)]

zipcodeCounts = filteredStores["Zip"].value_counts()

for zip in zipcodesToCount:
    count = zipcodeCounts.get(zip, 0) 
    print(str(zip)+": "+str(count))
```

Finally, we make a bar graph to visually see the number of convenience stores and supermarkets each zipcode has.

```python
zipcodeCounts.plot(kind='bar', color='pink', edgecolor='black')
```

## Conclusion:
Based on this data, the zipcode 15222 (The Strip District) has the most convenience stores and supermarkets. The second highest are 15206, which is East Liberty, 15212, which is Brighton Heights, and 15205, which is Robinson.

We decided to determine the "best" neighborhood based on how many of each component each zipcode has. Based on the stores component, the Strip District is the best.
